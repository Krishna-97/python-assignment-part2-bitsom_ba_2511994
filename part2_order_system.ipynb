# Task 2 — Cart Operations

```python
cart = []
#Logic for adding a items to cart
def add_to_cart(item_name, qty):
    if item_name not in menu:
        print(f"Error: '{item_name}' does not exist in menu.")
        return
    if not menu[item_name]['available']:
        print(f"Error: '{item_name}' is currently unavailable.")
        return
#Logic for update item quantity in the cart
    for entry in cart:
        if entry['item'] == item_name:
            entry['quantity'] += qty
            return
    cart.append({"item": item_name, "quantity": qty, "price": menu[item_name]['price']})
```

```python
# Logic for removing items from cart
def remove_from_cart(item_name):
    global cart
    original_len = len(cart)
    cart = [entry for entry in cart if entry['item'] != item_name]
    if len(cart) == original_len:
        print(f"Error: '{item_name}' not found in cart.")
```

```python
# Stimulation of cart function using below values:
add_to_cart("Paneer Tikka", 2); print(f"Added Paneer Tikka x2: {cart}")
add_to_cart("Gulab Jamun", 1); print(f"Added Gulab Jamun x1: {cart}")
add_to_cart("Paneer Tikka", 1); print(f"Updated Paneer Tikka: {cart}")
add_to_cart("Mystery Burger", 1)
add_to_cart("Chicken Wings", 1)
remove_from_cart("Gulab Jamun"); print(f"Removed Gulab Jamun: {cart}")
```

```python
# Displaying Final Order Summary:
print("\n========== Order Summary ==========")
subtotal = 0
for entry in cart:
    line_total = entry['quantity'] * entry['price']
    subtotal += line_total
    print(f"{entry['item']:<15} x{entry['quantity']}    ₹{line_total:>7.2f}")

gst = subtotal * 0.05
total_payable = subtotal + gst
print("-" * 34)
print(f"Subtotal:                ₹{subtotal:>7.2f}")
print(f"GST (5%):                ₹{gst:>7.2f}")
print(f"Total Payable:           ₹{total_payable:>7.2f}")
print("====================================")

```

# Task 4 — Daily Sales Log Analysis

```python
# Add new day
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

def analyze_sales():
    daily_revenue = {}
    item_counts = {}
    # calculating and displaying Revenue/day daywise
    for date, orders in sales_log.items():
        revenue = sum(order['total'] for order in orders)
        daily_revenue[date] = revenue

        for order in orders:
            for item in order['items']:
                item_counts[item] = item_counts.get(item, 0) + 1

    print("Revenue per Day:")
    for date, rev in daily_revenue.items():
        print(f"{date}: ₹{rev:.2f}")
    #Calculating best day with maximum revenue
    best_day = max(daily_revenue, key=daily_revenue.get)
    print(f"\nBest-selling day: {best_day} (₹{daily_revenue[best_day]:.2f})")
    #Calculating and displaying most ordered items
    most_ordered = max(item_counts, key=item_counts.get)
    print(f"Most ordered item (by frequency in orders): {most_ordered}")

analyze_sales()
#Displaying enumerated list of Orders:
print("\nNumbered List of All Orders:")
global_index = 1
for date, orders in sales_log.items():
    for order in orders:
        items_str = ", ".join(order['items'])
        print(f"{global_index}. [{date}] Order #{order['order_id']} — ₹{order['total']:.2f} — Items: {items_str}")
        global_index += 1
```
